## Generate Cities List

```python
from citipy import citipy as ct
import random as r
import matplotlib.pyplot as plt
import numpy as np
import json
import requests
import pandas as pd

from scipy import stats as st
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
def get_URL(cityname):
    result = ('https://maps.googleapis.com/maps/api/geocode/json?address={0}&key={1}').format(cityname, gkey)
    return result
```

```python
numCities = 50
lat = [r.random() * 140 - 60 for _ in range(numCities)]
lng = [r.random() * 360 - 180 for _ in range(numCities)]
```

```python
plt.scatter(lng,lat)
```

```python

act_city = []
act_lat = []
act_lng = []
act_state = []
act_country = []

for idx, x in enumerate(lng):
    y = lat[idx]
    c = ct.nearest_city(x,y)
    
    
    url = get_URL(c.city_name)
    result = requests.get(url).json()['results'][0]
    location = result['geometry']['location']
    act_lat.append(location['lat'])
    act_lng.append(location['lng'])
    
    state = ""
    country = ""
    
    try:
        city = result['address_components'][0]['long_name']
        for add_idx in range(1,len(result['address_components'])):
            if result['address_components'][add_idx]['types'][0] == 'country':
                country = result['address_components'][add_idx]['long_name']
            elif result['address_components'][add_idx]['types'][0] == 'administrative_area_level_1': 
                state = result['address_components'][add_idx]['long_name']
                
        print(f'{city}, {state}, {country}, {lng[idx]:0.3f} => {act_lng[idx]:0.3f} longitude  {lat[idx]:0.3f} => {act_lat[idx]:0.3f} latitude')
    
        act_city.append(city)
        act_state.append(state)
        act_country.append(country)
        
    except:
        print(f"Error {idx}")
    
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
city_df = pd.DataFrame({"city":act_city,
                       "state": act_state,
                       "country": act_country,
                       "latitude": act_lat,
                       "longitude": act_lng})

city_df = city_df.drop_duplicates().reset_index(drop = True)

city_df.to_csv("../output_data/city_list.csv")
```

```python
if city_file != "":
    city_df = pd.read_csv(city_file)
```

```python
#look up the weather data for each city
temps = []
humid = []
clouds = []
wind_speed = []
temp_max = []

for idx, row in city_df.iterrows():
    lat = row['latitude']
    long = row['longitude']
    
    x = get_weather(lat, long)
    temps.append(x['main']['temp'])
    humid.append(x['main']['humidity'])
    clouds.append(x['clouds']['all'])
    wind_speed.append(x['wind']['speed'])
    temp_max.append(x['main']['temp_max'])
    
city_df['humidity'] = humid
city_df['temp'] = temps
city_df['temp_max'] = temp_max
city_df['clouds'] = clouds
city_df['wind'] = wind_speed
```

## Linear Regression

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
north = city_df_100[city_df_100['latitude'] >=0]
south = city_df_100[city_df_100['latitude'] <0]
```

```python
n_lat = north['latitude']
s_lat = abs(south['latitude'])
```

```python
n_maxtemp = north['temp_max']
s_maxtemp = south['temp_max']
```

```python
(slope_n, intercept_n, pvalue_n, rvalue_n, sterr_n) = st.linregress(n_lat,n_maxtemp) 
n_x = range(0,90,5)
n_y = slope_n * n_x + intercept_n
```

```python
plt.scatter(n_lat, n_maxtemp)
plt.plot(n_x, n_y, marker = "" , color = 'g')
plt.grid(True)
plt.xlabel('Latitude (North)')
plt.ylabel('Max Temperture (F)')
plt.title('Max Temp in Northern Hemisphere',size=14,color = 'purple')
plt.xlim(0,90)
plt.ylim(0,120)
plt.show()
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
(slope_s, intercept_s, pvalue_s, rvalue_s, sterr_s) = st.linregress(s_lat,s_maxtemp) 
s_x = range(0,90,5)
s_y = slope_s * s_x + intercept_s
```

```python
plt.scatter(s_lat, s_maxtemp)
plt.plot(s_x, s_y, marker = "" , color = 'red')
plt.grid(True)
plt.xlabel('Latitude (South)')
plt.ylabel('Max Temperture (F)')
plt.title('Max Temp in Southern Hemisphere',size=14,color = 'blue')
plt.xlim(0,90)
plt.ylim(0,120)
plt.show()
```
